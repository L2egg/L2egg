```python
import pandas as pd
import numpy as np

from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.metrics import f1_score, confusion_matrix, precision_recall_curve, roc_curve
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Binarizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

dataset = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sub = pd.read_csv('submission.csv')
```

```python
def solve_nah(dataset):
    dataset['hour_bef_temperature'].fillna(dataset['hour_bef_temperature'].mean(), inplace=True)
    dataset['hour_bef_precipitation'].fillna(1, inplace=True)
    dataset['hour_bef_windspeed'].fillna(dataset['hour_bef_windspeed'].mean(), inplace=True)
    dataset['hour_bef_humidity'].fillna(dataset['hour_bef_humidity'].mean(), inplace=True)
    dataset['hour_bef_visibility'].fillna(dataset['hour_bef_visibility'].mean(), inplace=True)
    dataset['hour_bef_ozone'].fillna(dataset['hour_bef_ozone'].mean(), inplace=True)
    dataset['hour_bef_pm10'].fillna(dataset['hour_bef_pm10'].mean(), inplace=True)
    dataset['hour_bef_pm2.5'].fillna(dataset['hour_bef_pm2.5'].mean(), inplace=True)
```

```python
solve_nah(dataset)
solve_nah(test)
```

```python
from sklearn.model_selection import train_test_split

X = dataset.drop('count', axis=1)
y = dataset['count']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg = reg.fit(X_train, y_train)

pred = reg.predict(test)
pred
```

```python
sub['count'] = pred
```
