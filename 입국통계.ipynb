```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv('foreign_to_korea.csv', encoding='cp949')
dataset.head()
```

```python
dataset = dataset.drop(['국적별(1)','국적별(2)'], axis=1)
dataset.head()
```

```python
x = np.arange(31)
a = dataset.columns
a = list(a)
b = dataset.iloc[0,:]
b = list(b)
plt.figure(figsize=(15,6))
plt.bar(x, b)
plt.xticks(x, a, rotation=45)
plt.show()
```

```python
dataset = dataset.T
dataset.head()
```

```python
X = dataset.index.values
y = dataset.iloc[:,0].values
X, y
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
ax.scatter(x=X, y=y)
plt.xticks(rotation=45)
plt.grid()
plt.plot(X,y)
plt.show()
```

```python
dataset_tour = pd.read_csv('tour_data.csv', encoding='cp949')
dataset_tour = dataset_tour.drop('계정코드별', axis=1)
dataset_tour
```

```python
y_export = dataset_tour.iloc[0,:].values
X = dataset_tour.columns.values
X.shape, y_export.shape
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
ax.scatter(x=X, y=y_export)
plt.xticks(rotation=45)
plt.grid()
plt.plot(X,y_export)
plt.title('tour_revenue')
plt.xlabel('year')
plt.ylabel('million', rotation=0, loc='top')
plt.show()
```

```python
dataset_h1n1 = pd.read_csv('H1N1_data.csv')
dataset_h1n1 = dataset_h1n1.drop('Link', axis=1)
dataset_h1n1.tail()
```

```python
# 신종플루 한국 누적 확진자
df_kh = dataset_h1n1[dataset_h1n1['Country'] == 'Republic of Korea']
```

```python
X = df_kh.iloc[:,0].values
y = df_kh.iloc[:,2].values

fig, ax = plt.subplots(figsize=(10, 7))
ax.scatter(x=X, y=y)
plt.xticks(rotation=45)
plt.grid()
plt.plot(X,y)

plt.show()
```
